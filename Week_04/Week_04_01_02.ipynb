# Continuity

When is a function continouous?



Geometrically, you can think of a function that is continuous at every number in an
interval as a function whose graph has no break in it. The graph can be drawn without
removing your pen from the paper.

## Definition

function is **continuous at a number $a$** if

\begin{equation}
 \lim_{x\to a} f(x) = f(a)
\end{equation}

This deifinition requires three things:

1. $f(a)$ is defined (that is, $a$ is in the domain of $f$).
</br>

2. $lim_{x\to a} f(x)$ exists
</br>

3. $lim_{x\to a} f(x)=f(a)$ 

- A continuous function $f$ has the property that a small change in $x$ produces only a small change in $f$.
</br>


- The change in $f$ can be kept as small as we please by keeping the change in $x$ sufficiently small.


We say that $f$ is **discontinuous at $a$** (or has a **discontinuity** at $a$) if $f$ is not continuous at $a$.

Here are some more examples of discontinuity:



A function is **continuous from the right at a number $a$** if

\begin{equation}
 \lim_{x\to a^+} f(x) = f(a)
\end{equation}

and is **continuous from the left at $a$** if

\begin{equation}
 \lim_{x\to a^-} f(x) = f(a)
\end{equation}


A function is **continuous on an interval** if it is continuous at
every number in the interval. (If f is defined only on one side of an endpoint of the
interval, we understand _continuous_ at the endpoint to mean _continuous from the
right or continuous from the left_.)

If $f$ and $g$ are continuous at $a$ and $c$ is a constant, then the following functions are also continuous at $a$:

1. $f+g$
</br>

2. $f-g$
</br>

3. $cf$
</br>

4. $fg$
</br>

5. $  \Large{\frac{f}{g}}$ if $g(a)\ne 0$

## Theorem

The following types of functions are continuous at every number in their domains:

1. Polynomials 
</br>

2. Rational functions 
</br>

3. Root functions
</br>

4. Trigonometric functions 
</br>

5. Inverse trigonometric functions
</br>

6. Exponential functions 
</br>

7. Logarithmic functions

## Theorem

If $g$ is continuous at $a$ and $lim_{x\to a}g(x)=b$, then  $lim_{x\to a}f\big(g(x)\big)=f(b)$. In other words,

\begin{equation}
 \lim_{x\to a} f\big(g(x)\big) = f\big(\lim_{x\to a} g(x)\big)
\end{equation}

## Theorem

If $g$ is continuous at $a$ and $f$ is continuous at $g(a)$, then the composite function $f\circ g$ given by $(f\circ g) (x)=f\big(g(x)\big)$ is continuous at $a$.

In other words: a continuous function of a continuous function is a continuous function.

## THE INTERMEDIATE VALUE THEOREM

Suppose that $f$ is continuous on the
closed interval $[a,b]$ and let $N$ be any number between $f(a)$ and $f(b)$, where
$f(a)\ne f(b)$. Then there exists a number $c$ in $(a,b)$ such that $f(c)=N$.

# Continuity in $\Re^3$

A function $f$ of two variables is called **continuous at** $(a,b)$ if 


\begin{equation}
\lim_{(x,y)\to (a,b)} f(x,y) =f(a,b)
\end{equation}

We say $f$ is continuous on $D$ if is continuous at every point $(a,b)$ in $D$.

# The tangent problem

A tangent to a curve is a line that touches the curve.

Tangent to a circle:

Line that intersects the circle once and only once.



For more complicated curves this definition is inadequate:

## Example

Find an equation of the tangent line to the parabola $y=x^2$ at the point P=(1,1).

- We need the the slope of the line $m$ at (0.5,0.25).
- We need 2 points to calculate a slope.
- We can compute an approximation to $m$ by choosing a nerby point $Q=(x,x^2)$

\begin{equation}
m_{PQ}=\frac{x^2-0.25}{x-0.5}
\end{equation}

- The line defined by the points $P$ and $Q$ is the _secant line_

```python
def tangent(px,qx,zoom):
    x=np.linspace(-10.2,10.2,1000)
    msec=(qx**2-px**2)/(qx-px)
    m=2*px

    plt.plot(x,x**2)

    if zoom:
        plt.ylim(-0.1,2)
        plt.xlim(0,2)     
    else:
        plt.ylim(-0.1,2)
        plt.xlim(-1,2)
        
    plt.axvline(x = 0, color = 'k')
    plt.axhline(y = 0, color = 'k')
    plt.plot(px,px**2,'o')
    plt.annotate('$P({},{})$'.format(px,round(px**2,1)),xy=(px+0.1,px**2), color='tab:orange')
    plt.plot(x,msec*(x-px)+px**2, '--', color='tab:orange')
    plt.plot(x,m*(x-px)+px**2)

    plt.plot(qx,qx**2,'o',color='tab:orange')
    plt.annotate('$Q({},{})$'.format(qx,round(qx**2,1)),xy=(qx-0.3,qx**2+0.1), color='tab:orange')
    
```

```python
tangent(px=0.5,qx=0.55,zoom=False)
```

We say that the slope of the tangent line is the limit of the slopes of the secant lines,
and we express this symbolically by writing

\begin{equation}
\lim_{Q\to P} m_{PQ} = m
\end{equation}

Then if $P=(0.5,0.25)$,

\begin{equation}
m=\lim_{x\to 0.5} m_{PQ} = \lim_{x\to 0.5}\frac{x^2-0.25}{x-0.5}=1
\end{equation}

We use the point-slope form of the equation of a line to write the equation of the tangent line through as


\begin{equation}
y=1(x-0.5)+0.25
\end{equation}
