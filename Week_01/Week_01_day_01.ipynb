# Functions

A function is a **rule** that assigns to each element $x$ in a set $D$ exactly one element, called $y$, in a set $E$.
$f : x \in D \rightarrow y \in E$.



- Input: independent variable(s): $x$

- Output: dependent variable: $y$

- $D:$ domain

- $E:$ range (the set of all possible values of $f(x)$ as $x$ varies throughout the domain)

- Usually $D\in \Re$ and $E\in \Re$

## Algebraic representations

- Explicit form $y=f(x)$



- Parametric form    $\left\lbrace
       \begin{array}{ll}
         x(t) \\
         y(t) 
      \end{array}
  \right\rbrace$
  


- Implicit form $F(x,y)=0$

**Example 1:**

A straight line

- Explicit representation: $y=x+2$

- Implicit representation: $y−x−2=0$

- Parametric equation: $\left\lbrace
       \begin{array}{ll}
         x=t \\
         y=t+2 
      \end{array}
  \right\rbrace$ (there are ∞ many ways) 

## Graphic representations

**Arrow diagram**



**Graph**

get the same information from both without ambiguities.

**THE VERTICAL LINE TEST** 

A curve in the xy-plane is the graph of a function if
and only if no vertical line intersects the curve more than once.

## Simmetry

If a function satisfies $f(-x)=f(x)$ for every number in its domain, then is called an **even** function.



For instance $f(x)=x^2$

If $f$ satisfies $f(-x)=-f(x)$ for every number in its domain, then is called an **odd** function.



For instance $f(x)=x^3$

## Increasing and Decrasing functions

- A function is called **increasing** on an interval $I$ if $f(x_1)<f(x_2)$ whenever $x_1<x_2$ in $I$.

- It is called **decreasing** on if $f(x_1)>f(x_2)$ whenever $x_1<x_2$ in $I$.

- The domain of any polynomial is $(-\infty,\infty)$. If the leading coefficient $a_n\ne0$, then the degree of the polynomial is $n$.

## Basic operations: New functions from old functions

**Vertical and horizontal shifts**
- $f (x−a)$ : translation to the right (delay)
- $f (x+a)$ : translation to the left 
-  $y+b= f ( x)$ : translation upward
-  $y−b= f ( x)$ : translation downward

**Vertical and horizontal stretching and reflecting**

Suppose $c>1$
- $y=cf(x)$ : stretch the graph vertically by a factor of c.
- $y=(1/c)f(x)$ : compress the graph vertically by a factor of c.
- $y=f(cx)$ : compress the graph horizontally by a factor of c.
- $y=f(x/c)$ : stretch the graph horizontally by a factor of c.
- $f (−x)$ : reflection about the y-axis.
- − $f (x)$ : reflection about x-axis
- The inverse: $f^ {-1}(x)$ reflection about $y=x$.

Let's plot some transformations...

## The inverse of a function

The inverse of a function is the operation that does just the opposite of the original one. 
In other words, the inverse **undoes** what the function have done before.

**Definition**

Let $f$ be a one-to-one function with domain and range . Then its **inverse function** has domain and range and is defined by 

\begin{equation}
f^{-1}=x \Leftrightarrow f(x)=y
\end{equation}

for any $y$ in $B$.

Then ...

- domain of $f^{-1}$ = range of $f$
- range of $f^{-1}$ = domain of $f$

**Definition**

A function is called a one-to-one function if it never takes on the same value twice; that is,

\begin{equation}
f(x_1)\ne f(x_2),~whenever~x_1\ne x_2
\end{equation}

**Horizontal line test**




- $f^{-1}\big(f(x)\big)=x$ for every $x\in A$ (domain of $f$)
- $f\big(f^{-1}(x)\big)=x$ for every $x\in B$ (range of $f$)

**How to find the inverse of a function $f$**

1. Write $y=f(x)$.
2. Solve this equation for $x$ in terms of $y$.
3. interchage $x$ and $y$.

## Combination of functions

Given $f(x)$ and $g(x)$ with domains $A$ and $B$, respectively:

- $(f+g)(x)=f(x)+g(x)$, domain: $A\cap B$
- $(f-g)(x)=f(x)-g(x)$, domain: $A\cap B$
- $(fg)(x)=f(x)g(x)$, domain: $A\cap B$
- $(f/g)(x)=f(x)/g(x)$, domain: $\{x\in A\cap B|g(x)\ne0\}$

## Exponential Functions

$y(x)=a^x$, the variable $x$ is the exponent, $a>0$ and $x \in \Re$.

```python
x=np.linspace(-1.5,1.0)

def exp_func(a,x):
    return a**x
def plot_exps():
    plt.plot(x,exp_func(0.5,x))
    plt.plot(x,exp_func(0.25,x))
    plt.plot(x,exp_func(10,x))
    plt.plot(x,exp_func(4,x))
    plt.plot(x,exp_func(2,x))
    # plt.plot(x,exp_func(1.5,x))
    plt.plot(x,exp_func(1,x),'--')
    plt.xlim([-1,1])
    plt.ylim([-0.2,3])
    plt.axhline(y = 0, color = 'k', linestyle = '-')
    plt.axvline(x = 0, color = 'k', linestyle = '-')

    plt.text(-0.6, 2.5, '$a=1/4$')
    plt.text(-0.6, 1.6, '$a=1/2$')
    plt.text(-0.6, 1.05, '$a=1$')
    plt.text(0.45, 2.5, '$a=10$')
    plt.text(0.45, 1.7, '$a=4$')
    plt.text(0.45, 1.25, '$a=2$')
```

```python
plot_exps()
```

Properties:

- Domain: $(-\infty,\infty)$.
- In the previous graph, all of the curves pass through the same point (0,1) since $a^{0}=1$.
- As the base $a$ gets larger, the exponential function grows more rapidly (for x>0).
- There are basically three kinds of exponential functions:
    - if $a=1$, it is a constant.
    - if $0<a<1$, the exponential function decreases.
    - if $0>a$, the exponential function increases.
    - if $a\ne 1$, the range is: $(0,\infty)$.

**Laws of exponents**

- $a^{x+y}=a^xa^y$

- $a^{x-y}=a^x/a^y$

- $(a^x)^y=a^{xy}$

- $(ab)^x=a^xb^x$

Proof: Appendix G from Stewart - Calculus - Early Transcendentals (Thomson, 2008).

# Logarithmic Function

**Definition**

\begin{equation}
log_ax=y \Leftrightarrow a^y=x, ~ if a>0~and~a\ne1
\end{equation}

- $log_a(a^x)=x$ for every $x \in \Re$
- $a^{log_ax}=x$ for every $x>0$
- It is the inverse function to exponentiation. 
- The logarithm of a number x to the base a, $\log_a(x)$, is the exponent to which b must be raised, to produce x.
- The logarithm of base $e$ is the $natural~logarithm$, $\ln(x)$.
- $x>0,~a>0$

Properties:

- $\log_{a}\big(xy\big)=\log_bx+\log_by$

- $\log_a\big(\frac{x}{y}\big)=\log_bx-\log_by$

- $\log_a\big(x^p)=p~\log_ax$

- $\log_a\sqrt[p] x=\frac{\log_ax}{p}$

Change of base $b\rightarrow k$:

\begin{equation}
\log_kx=\frac{\log_bx}{\log_bk}
\end{equation}

```python
t=np.linspace(0.01,2,1000)

plt.plot(t,np.log(t))
plt.axhline(y = 0, color = 'k', linestyle = '-')
plt.axvline(x = 0, color = 'k', linestyle = '-')
```

Domain: $\{x|0< x\}= (0,+\infty)$

Range: $\{y|~ y \in \Re\}$

## Power functions

\begin{equation}
f(x)=X^a, ~where ~a ~is ~a ~consant.
\end{equation}

- When $a=1,2,3,...n$, they are polynomials
- If $a=1/n$ where $n$ is a positive integer, the function is a **root function**.

```python
x=np.linspace(0,1,1000)

plt.plot(x,x**(1/2))
plt.plot(x,x**(1/4))
plt.plot(x,x**(1/6))
plt.axhline(y = 0, color = 'k', linestyle = '-')
plt.axvline(x = 0, color = 'k', linestyle = '-')
# plt.legend(['exp','log'])
```

```python
x=np.linspace(-1,1,1000)
def root(x,n):
    result = []
    for val in x:
        if val > 0:
            result.append(val ** (1./n))
        elif val < 0:
            result.append(-np.abs(val) ** (1./n))
    return result
            
plt.plot(x,root(x,3))
plt.plot(x,root(x,5))
plt.plot(x,root(x,7))
plt.axhline(y = 0, color = 'k', linestyle = '-')
plt.axvline(x = 0, color = 'k', linestyle = '-')
# plt.legend(['exp','log'])
```

- If $a=-1$, $f(x)$ is a reciprocal function.

## Algebraic Functions

A function is called an **algebraic function** if it can be constructed using algebraic operations
(such as addition, subtraction, multiplication, division, and taking roots) starting
with polynomials. Any rational function is automatically an algebraic function. 

Examples:

\begin{equation}
f(x)=\sqrt{x^2+1}
\end{equation}

\begin{equation}
f(x)=\frac{x^4-16x^2}{x+\sqrt{x}}+(x-2)^3\sqrt[3]{x+1}
\end{equation}
