# The Chain Rule

Suppose you are asked to differentiate the function

\begin{equation}
F(x) = \sqrt{x^2+1}
\end{equation}

The differentiation formulas you learned in the previous sections do not enable you to calculate $F'(x)$.

- Observe that $F(x)$ is a composite function: $F(x)=f(x)\circ g(x)$ where $f(x)=\sqrt{x}$ and $g(x)=x^2+1$.
- how to differentiate both $f$ and $g$, so it would be useful to have a rule that tells us how to find the derivative of $F=f\circ g$.

## THE CHAIN RULE

If $g$ is differentiable at $x$ and $f$ is differentiable at $g(x)$, then the
composite function $F = f\circ g$ defined by $F(x)=f\big(g(x)\big)$ is differentiable at $x$ and
is given by the product

\begin{equation}
F'(x) = f'\big(g(x)\big)\cdot g'(x)
\end{equation}

In Leibniz notation, 


\begin{equation}
\frac{dF}{dx} =\frac{df}{dg} \frac{dg}{dx}
\end{equation}

**This is one of the most important of the differentiation rules.**

# Parametric diferentiation

Consider the following parametric equation:

\begin{equation}
\left\lbrace
       \begin{array}{ll}
         y=F(t) \\
        x=x(t)
      \end{array}
  \right.
\end{equation}

Then, $y=F\big(x(t)\big)$ and

\begin{equation}
\frac{dy}{dt}=\frac{dy}{dx}\frac{dx}{dt}\Rightarrow\frac{dy}{dx}=\frac{\frac{dy}{dt}}{\frac{dx}{dt}}
\end{equation}

## Example

Given 

\begin{equation}
\left\lbrace
       \begin{array}{ll}
         y=t^2+5 \\
        x=e^t
      \end{array}
  \right.
\end{equation}

- slope on the graph y respect to t: 

\begin{equation}
\frac{dy}{dt}=2t
\end{equation}

- slope on the graph x respect to t:

\begin{equation}
\frac{dx}{dt}=e^t
\end{equation}

- slope on the graph y respect to x:

\begin{equation}
\frac{dy}{dx}=\frac{2t}{e^t}
\end{equation}

# The tangent problem in $\Re^2$

Consider the problem of finding the tangent line to a curve $y=f(x)$ at a given point $P (x_0,y_0)$. The equation of a line that crosses the point $P$ is

\begin{equation}
 y-y_0=m(x-x_0), ~~ where ~~ m=f'(x_0)
\end{equation}

## Example

Find the tangent line to the curve $y = x^3-x$ at $x=1/2$.

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-2,2,1000)

plt.plot(x,x**3-x)
plt.plot(x,-0.25*(x-0.5)-3/8)
plt.plot(0.5,-3/8,'o')
plt.axvline(x = 0, color = 'k')
plt.axhline(y = 0, color = 'k')
plt.ylim(-1,1)
plt.xlim(-1,2)
```

Special cases:

- If the tangent line is horizontal, $m = 0$.
- If the tangent line is vertical, $m = \infty$.
- If two straight lines are parallel, their slope will be the same.
- If two straight lines are normal or perpendicular, their slopes will be related as

\begin{equation}
m_1\perp m_2 \Rightarrow m_1 = \frac{-1}{m_2}
\end{equation}

# L'Hopital Rule

Suppose $f$ and $g$ are differentiable and $g'(x)\ne 0$ on an open interval $I$ that contains $a$ (except possibly at $a$). Suppose that

\begin{equation}
\lim_{x\to a} f(x) =0 ~~ and ~~ \lim_{x\to a} q(x) =0
\end{equation}

or that 

\begin{equation}
\lim_{x\to a} f(x) =+\pm \infty ~~ and ~~ \lim_{x\to a} q(x) =+\pm \infty
\end{equation}

(In other words, we have an indeterminate form of type $\frac{0}{0}$ or $ \frac{\infty}{\infty} $. Then 



\begin{equation}
\lim_{x\to a} \frac{f(x)}{g(x)} =\lim_{x\to a} \frac{f'(x)}{g'(x)}
\end{equation}
 
if the limit on the right side exists (or is $\infty$ or $-\infty$).

- **Note 1** L’Hospital’s Rule says that the limit of a quotient of functions is equal to the limit of the quotient of their derivatives, provided that the given conditions are satisfied. It is especially important to verify the conditions regarding the limits of $f$ and $g$ before using l’Hospital’s Rule.

- **Note 2** L’Hospital’s Rule is also valid for one-sided limits and for limits at infinity or negative infinity; that is,  $x\to a$” can be replaced by any of the symbols $x\to a^+$, $x\to a^-$, $x\to \infty$, or $x\to -\infty$.

- **Note 3** For the special case in which $f(a)=g(a)=0$, $f'$ and  $g'$ are continuous, and $g'(a)\ne 0$, it is easy to see why l’Hospital’s Rule is true. In fact, using the alternative form of the definition of a derivative.

\begin{equation}
\lim_{x\to a} \frac{f'(x)}{g'(x)} =\frac{f'(a)}{g'(a)}=\frac{\lim_{x\to a} \frac{f(x)-f(a)}{x-a}}{\lim_{x\to a} \frac{g(x)-g(a)}{x-a}}=\lim_{x\to a}\frac{\frac{f(x)-f(a)}{x-a}}{\frac{g(x)-g(a)}{x-a}}=\lim_{x\to a}\frac{f(x)-f(a)}{g(x)-g(a)}=lim_{x\to a}\frac{f(x)}{g(x)}
\end{equation}
