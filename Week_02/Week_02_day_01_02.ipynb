# Parametric curves

In the previous graph,

- The curve $C$ is not a function. It is imposible to decribe it as $y=f(x)$.
- But $x$ and $y$ are functions of the parameter $t$.

\begin{equation}
 \left\lbrace
       \begin{array}{ll}
         x=f(t) \\
         y=g(t) 
      \end{array}
  \right\rbrace
\end{equation}

- As $t$ varies, the point $(x,y)=\big(f(t),g(t)\big)$ varies and traces out a curve $C$.

## Equation for a Circle

Circle of unit radious:

\begin{equation}
 \left\lbrace
       \begin{array}{ll}
         x=\cos(t)  \\
         y=\sin(t) 
      \end{array}
  \right.
 ,~0\le t \le 2\pi~~ \Rightarrow x^2+y^2=1
\end{equation}

As $t$ increases from $0$ to $2\pi$, the point $(x,y)=(\cos t, \sin t)$ moves once around the circle in
the counterclockwise direction starting from the point $(0,1)$.


```python
import numpy as np
import matplotlib.pyplot as plt


t=np.linspace(0,2*np.pi)

plt.plot(np.cos(t),np.sin(t))
plt.axhline(y = 0, color = 'k', linestyle = '-')
plt.axvline(x = 0, color = 'k', linestyle = '-')

ax = plt.gca()
ax.set_aspect('equal')
```

**Question:** What happens if we change $t$ by $2t$?

```python
# Enable interactive plot
%matplotlib notebook

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()

t=np.linspace(0,2*np.pi)

ax.plot(np.cos(t),np.sin(t))
ax.plot(0,0,'*')
ax = plt.gca()
ax.set_aspect('equal')

line2, = ax.plot([],'X',markersize=10)     # A tuple unpacking to unpack the only plot
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)

def animate(t):
    x = np.cos(2*np.pi * t/100)
    y = np.sin(2*np.pi * t/100)
    
    x1 = np.cos(2*np.pi * 2*t/100)
    y1 = np.sin(2*np.pi * 2*t/100)
    
    line2.set_data(([x,x1], [y,y1]))
    return line2

anim = FuncAnimation(fig, animate, frames=100, interval=20)
plt.show()
```

- Different sets of parametric equations can represent the same curve.
- Thus we distinguish between a **curve**, which is a set of points, and a **parametric curve**, in which the points are traced in a particular way.

## General circle

Parametric equations for the circle with center $(h,k)$ and radius $r$:

- Set of points $(x,y)$ that satisfies: $\sqrt{(x-h)^2+(y-k)^2}=r ~~\Rightarrow ~~ (x-h)^2+(y-k)^2 = r^2$




- A parametric representation may be given by:


\begin{equation}
 \left\lbrace
       \begin{array}{ll}
         x-h=r\cos(\omega t)  \\
         y-k=r\sin(\omega t) 
      \end{array}
  \right.~~~~~
 \Rightarrow ~~~~~
  \left\lbrace
       \begin{array}{ll}
         x=r\cos(\omega t) +h  \\
         y=r\sin(\omega t) +k
      \end{array}
  \right.,~~~~~0\le t \le T
\end{equation}

- $T=2\pi/\omega$

```python
t=np.linspace(0,2*np.pi)

r=1.5
k=1
h=0.5

plt.plot(np.cos(t),np.sin(t))
plt.plot(r*np.cos(t)+h,r*np.sin(t)+k)
plt.axhline(y = 0, color = 'k', linestyle = '-')
plt.axvline(x = 0, color = 'k', linestyle = '-')
plt.plot(h, k, marker="X", markersize=10, color="tab:orange")
plt.title('Circle of radious r and center (h,k)')
ax = plt.gca()
ax.set_aspect('equal')
```

What is the domain of a general circle?

- You can obtain the domain from the graph.
- Or from the study of its explicit functions.
\begin{equation}
\sqrt{(x-h)^2+(y-k)^2}=r ~~\Rightarrow ~~ y=k\pm\sqrt{r^2-(x-h)^2}
\end{equation}


\begin{equation}
\Rightarrow r^2-(x-h)^2>0 \Rightarrow |r|>|x-h|
\end{equation}

**Observation**

Completing the square can be used as well in order to identify the radius or a circle and its center.

Ex: $x^2-2x+y^2=0$ is a circle.


**Example:**

Which curve is:

\begin{equation}
 \left\lbrace
       \begin{array}{ll}
         x=\sin(t)  \\
         y=\sin^2(t) 
      \end{array}
  \right.
\end{equation}


```python
# Enable interactive plot
%matplotlib notebook

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()

t=np.linspace(0,2*np.pi)

ax.plot(np.sin(t),np.sin(t)**2)
ax = plt.gca()
ax.set_aspect('equal')

line2, = ax.plot([],'X',markersize=10)     # A tuple unpacking to unpack the only plot
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)

def animate(t):
    x = np.sin(2*np.pi *t/100)
    y = np.sin(2*np.pi *t/100)**2
    line2.set_data((x, y))
    return line2

anim = FuncAnimation(fig, animate, frames=100, interval=20)
plt.show()
```

**Example:**

Graph the curve $x=y^4-3y^2$.

- You can solve the equation for y as four functions of x and graph them individually.
- You can transform the expression into a parametric equation:

\begin{equation}
 \left\lbrace
       \begin{array}{ll}
         x=t^4-3t^2  \\
         y=t
      \end{array}
  \right.
\end{equation}

```python
t=np.linspace(-3,3)

plt.plot(t**4-3*t**2,t)
plt.axhline(y = 0, color = 'k', linestyle = '-')
plt.axvline(x = 0, color = 'k', linestyle = '-')
plt.xlim([-4,4])
plt.ylim([-4,4])
ax = plt.gca()
ax.set_aspect('equal')
```

In General ...

- If we need to graph $x=g(y)$, we can use

\begin{equation}
 \left\lbrace
       \begin{array}{ll}
         x=g(t)  \\
         y=t
      \end{array}
  \right.
\end{equation}

- If we need to graph $y=f(x)$, we can use

\begin{equation}
 \left\lbrace
       \begin{array}{ll}
         x=t \\
         y=f(t)
      \end{array}
  \right.
\end{equation}

## More examples of parametric curves...

```python
# Enable interactive plot
%matplotlib notebook

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()

t=np.linspace(-10*np.pi,10*np.pi,1000)

ax.plot(t+2*np.sin(2*t),t+2*np.cos(5*t))
ax = plt.gca()
ax.set_aspect('equal')

line2, = ax.plot([],'X',markersize=10)     # A tuple unpacking to unpack the only plot
ax.set_xlim(-6.5,6.5)
ax.set_ylim(-8, 8)

def animate(t):
    t=-np.pi+t/50
    x = t+2*np.sin(2*t)
    y = t+2*np.cos(5*t)
    line2.set_data((x, y))
    return line2

anim = FuncAnimation(fig, animate, frames=1000, interval=20)
plt.show()
```

# Piecewise Functions

## The absolute value function

\begin{equation}
 f(x)=|x|=\sqrt{x^2}=\left\lbrace
       \begin{array}{ll}
         -x, ~~x<0 \\
         x,~~x\ge0 
      \end{array}
  \right.
\end{equation}

```python
x1=np.linspace(0,3)
x2=np.linspace(-3,0)

plt.plot(x1,x1)
plt.plot(x2,-x2, color='tab:blue')

plt.axhline(y = 0, color = 'k', linestyle = '-')
plt.axvline(x = 0, color = 'k', linestyle = '-')
plt.xlim([-2,2])
plt.ylim([-2,2])
ax = plt.gca()
ax.set_aspect('equal')
```
